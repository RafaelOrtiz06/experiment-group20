```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
stress_data = pd.read_csv('group20-experiment.csv')
```

```python
stress_data.head()
```

```python
# Count all the rows
row_count = stress_data.shape[0]

# Print the row count
print("Total number of rows:", row_count)
```

```python
# Calculate the count of error responses (response code != 200)
error_count = len(stress_data[stress_data['responseCode'] != 200])

# Calculate the count of successful responses (response code == 200)
success_count = len(stress_data[stress_data['responseCode'] == 200])

# Create a bar graph
categories = ['Error', 'OK']
counts = [error_count, success_count]

plt.bar(categories, counts, color=['red', 'green'])

plt.xlabel('Response')
plt.ylabel('Count')
plt.title('Responses')
plt.show()
```

```python
# Extract 'elapsed' and 'timeStamp' columns (assuming 'timeStamp' exists)
elapsed_time = stress_data['elapsed']
timestamp = stress_data['timeStamp']  # If 'timeStamp' is available, use it; otherwise, use row number

# Create a line graph
plt.plot(timestamp, elapsed_time)
plt.xlabel('Timestamp or Row Number')
plt.ylabel('Elapsed Time (ms)')
plt.title('Elapsed Time Over Time')
plt.grid(True)

# Show the plot
plt.show()
```
